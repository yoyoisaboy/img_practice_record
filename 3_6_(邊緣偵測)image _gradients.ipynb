```python
import cv2
import numpy as np
import sys
from matplotlib import pyplot as plt

#影像梯度計算，色彩或強度方向性變化

img_path = "data/opencv.png"
img = cv2.imread(img_path)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#影像中物體與物體或物體與背景的交界線
def lapacian(img_path): 
	
	img = cv2.imread(img_path)

	if img is None:
		sys.exit("無法讀取影像...")

	#laplacian
	laplacian = cv2.Laplacian(img,cv2.CV_64F)
    
	plt.subplot(121),plt.imshow(img),plt.title('Original')
	plt.xticks([]), plt.yticks([])
	plt.subplot(122),plt.imshow(laplacian),plt.title('lapacian')
	plt.xticks([]), plt.yticks([])
	plt.show()
	#print(laplacian.shape)
	#cv2.imshow('laplacian',laplacian)
	#cv2.imshow('src',img)
	
	#k = cv2.waitKey(0)
	#cv2.destroyAllWindows() 
#lapacian
lapacian(img_path)
```

```python
def Sobel(img_path): 
	
	img = cv2.imread(img_path)

	if img is None:
		sys.exit("無法讀取影像...")
	#sobelx
	sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
	#sobely
	sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)

	plt.subplot(121),plt.imshow(sobelx),plt.title('sobelx')
	plt.xticks([]), plt.yticks([])
	plt.subplot(122),plt.imshow(sobely),plt.title('sobely')
	plt.xticks([]), plt.yticks([])
	plt.show()

	#cv2.imshow('sobelx',sobelx)
	#cv2.imshow('sobely',sobely)
	
	#k = cv2.waitKey(0)
	#cv2.destroyAllWindows()
#Sobel
Sobel(img_path)
```

```python
def Canny(img_path): 

    img = cv2.imread(img_path)

    if img is None:
        sys.exit("無法讀取影像...")
    canny = cv2.Canny(img, 30, 150)
    plt.subplot(122),plt.imshow(canny),plt.title('canny')
    plt.xticks([]), plt.yticks([])
    plt.show()


    #cv2.imshow('canny',canny)
    #k = cv2.waitKey(0)
    #cv2.destroyAllWindows() 
    #Canny
Canny(img_path)
```
